```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import pymysql
import re
from dotenv import load_dotenv
load_dotenv()
import os
```

```python
db_connection_str = os.getenv("DB_CONNECTION_STR")

db_connection = create_engine(db_connection_str)

df = pd.read_sql('SELECT * FROM crash_db_Chicago', con=db_connection)
df = df.replace('',np.nan)
```

```python
df2 = pd.DataFrame()
df2['cid'] = df.cid
df2['city'] = df.dataFrom
df2['crash_time'] = pd.to_datetime(df.CRASH_DATE)
df2['crash_date'] = pd.to_datetime(df2['crash_time'].dt.date)
```
