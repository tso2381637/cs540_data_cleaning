```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import pymysql
import re
from dotenv import load_dotenv
load_dotenv()
import os
```

```python
db_connection_str = os.getenv("DB_CONNECTION_STR")

db_connection = create_engine(db_connection_str)

df = pd.read_sql('SELECT * FROM crash_db_Austin', con=db_connection)
df = df.replace('',np.nan)
```

```python
df2 = pd.DataFrame()
df.head(10)
```

```python
# 
df2['city'] = df['dataFrom']

df2[['latitude','longitude']] = df[['latitude','longitude']].astype('float64')

print(df2.info())

df2.describe()
```

```python
df2['latitude'] = df2['latitude'].fillna(df2['latitude'].mean())
df2['longitude'] = df2['longitude'].fillna(df2['longitude'].mean())

df2[['latitude','longitude']].describe()
```

```python
df2['cid'] = df['cid']
df2['street_name'] = df.street_name
df2['injured_count'] = df.tot_injry_cnt
```

```python
df2['crash_time'] = df['crash_date'] + ' ' + df['crash_time']

df2.crash_time = pd.to_datetime(df2.crash_time)
df2['crash_date'] = pd.to_datetime(df2['crash_time'].dt.date)
df2.head()
```

```python
df2.to_sql('cleaned_crash_Austin',con = db_connection, if_exists="replace",index=False)
```
