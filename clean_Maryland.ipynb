```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import pymysql
import re
from dotenv import load_dotenv
load_dotenv()
import os
```

```python
db_connection_str = os.getenv("DB_CONNECTION_STR")

db_connection = create_engine(db_connection_str)
```

```python
df = pd.read_sql('SELECT * FROM crash_db_Maryland', con=db_connection)
df = df.replace('',np.nan)

df2 = pd.DataFrame()
df.head(10)
```

```python
print(df.shape)
df.columns
```

```python
df.info()
```

```python
df2['cid'] = df.cid
df2['latitude'] = df.LATITUDE
df2['longitude'] = df.LONGITUDE
df2['city'] = df.dataFrom
df2['street_name'] = df.MAINROAD_NAME
```

```python
df[['ACC_TIME','ACC_DATE']].dtypes
```

```python
df2['crash_time'] = df.ACC_DATE + ' ' + df.ACC_TIME
df2['crash_time'] = pd.to_datetime(df2['crash_time'])
df2['crash_date'] = pd.to_datetime(df2['crash_time'].dt.date)
```

```python
df2['injured_count'] = df['REPORT_TYPE'].replace({'Property Damage Crash':0,'Fatal Crash':1,'Injury Crash':1})
```
