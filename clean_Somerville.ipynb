```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import pymysql
import re
from dotenv import load_dotenv
load_dotenv()
import os
```

```python
db_connection_str = os.getenv("DB_CONNECTION_STR")

db_connection = create_engine(db_connection_str)

df = pd.read_sql('SELECT * FROM crash_db_Somerville', con=db_connection)
df = df.replace('',np.nan)
```

```python
df2 = pd.DataFrame()
df.head(10)
```

```python
df.shape
```

```python
df.columns
```

```python
df2['city'] = df['dataFrom']

df2['cid'] = df['cid']

df2['injured_count'] = df['Injury']
```

```python
df2['crash_time'] = pd.to_datetime(df['Date'])
df2['crash_date'] = pd.to_datetime(df2['crash_time'].dt.date)
```

```python
df2.injured_count = df2.injured_count.fillna('0')

df2.injured_count = df2.injured_count.astype('str').astype('int')
```

```python
df2['longitude'] = 0
df2['latitude'] = 0
```

```python
df2['street_name'] = df['Location']
```

```python
df2.head()
```

```python
df2.to_sql('cleaned_crash_Somerville',con = db_connection, if_exists="replace",index=False)

df2.to_sql('home_allcleanedcrashdata',con = db_connection, if_exists="append",index=False)
```
